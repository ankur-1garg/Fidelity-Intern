```python
birth_years = np.array(df['driver_age_raw'])
current_year = 2025
new_ages = np.where(
    ~np.isnan(birth_years),  # Check for non-NaN values
    current_year - birth_years,  # Calculate age if birth year exists
    np.nan  # Keep NaN for missing birth years
)
updated_ages = np.array(new_ages)

updated_ages
```

```python
df.head()
```

```python
df['updated_age_2025'] = updated_ages  # Add new column with updated ages
df['updated_age_2025'] = df['updated_age_2025'].round().astype('Int64')  # Int64 allows NaN values
df.to_csv('police.csv', index=False)
```
